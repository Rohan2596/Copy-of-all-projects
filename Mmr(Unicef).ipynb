```python
#import Data set
import pandas as pd
url="Desktop/Juypter/MMR_Matdeaths_LTR-trend-estimates-1990-2015_MMEIG-1.xlsx"
mmr=pd.read_excel(url)
mmr.head()
```

```python
#Dropping all unwanted data set points 
mmr.drop(columns=['Unnamed: 10','Unnamed: 9','Unnamed: 8'],inplace=True)
mmr.drop([0,1,2,3,],inplace=True)
```

```python
#renaming dataset for further analysis and 
mmr.rename(columns={'Trends in estimates of maternal mortality ratio (MMR; maternal deaths per 100,000 live births) 1990-2015':'MMR(1990-2015)ISOC','Unnamed: 1':'Countries/Regions', 'Unnamed: 2':'1990', 'Unnamed: 3':'1995', 'Unnamed: 4':'2000', 'Unnamed: 5':'2005','Unnamed: 6':'2010','Unnamed: 7':'2015'},inplace=True)
```

```python
#using algorithms for making prediction mmr rates
y=mmr['2015']
x_features=['1990','1995','2000','2005','2010']
x=mmr[x_features]
print(x.head())
print(x.info())
```

```python
from sklearn.linear_model import LinearRegression
win_model=LinearRegression()
win_model.fit(x,y)
print(win_model.predict(x.head()))
```

```python
mmr['2015_1']=win_model.predict(x)
mmr.head()
```

```python
#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
predicted_win_model=win_model.predict(x)
mean_absolute_error(y,predicted_win_model)
```

```python
#Importing Algorithms
from sklearn.tree import DecisionTreeRegressor 
win_model1=DecisionTreeRegressor(random_state=1)
win_model1.fit(x,y)
print("making Predictionsfor the following 5 Purchase")
print(x.head()) 
print("The predictions are")
print(win_model1.predict(x.head()))
```

```python
mmr['2015_2']=win_model1.predict(x)
mmr.head()
#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
predicted_win_model1=win_model1.predict(x)
mean_absolute_error(y,predicted_win_model1)
```
