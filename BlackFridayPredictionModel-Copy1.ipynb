```python
#NOT Taking into cansidering Martial status
y=bf.Purchase
bf_features=['OCC','PC1','PC2','PC3',]
x=bf[bf_features]
x.describe()
x.head()
```

```python
#Importing Algorithms
from sklearn.tree import DecisionTreeRegressor 
bf_model=DecisionTreeRegressor(random_state=1)
bf_model.fit(x,y)
print("making Predictionsfor the following 5 Purchase")
print(x.head()) 
print("The predictions are")
print(bf_model.predict(x.head()))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
train_X,val_X,train_y,val_y=train_test_split(x,y,random_state=1)
bf_model1=DecisionTreeRegressor(random_state=1)
bf_model.fit(train_X,train_y)
bf_model_predictions=bf_model.predict(val_X)
val_mae=mean_absolute_error(bf_model_predictions,val_y)
print("Validation MAE:{:,.0f}".format(val_mae))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000,50000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
